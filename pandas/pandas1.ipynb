```python
# 열 하나일 때, Series

series = pd.Series([70, 60, 90], index=["국어","영어","수학"])
series
```

```python
series.국어
print(series.국어)
print(series.index)
print(series.values)
print(series.shape) # 튜플 (3, ) : 3행
print(series.head(1))
print(series.tail(1))
```

```python
# 수정
# 인덱스 수정
series.index = ["미술", "음악","체육"]
series

# dtype : int64 -> int 64비트 표현
# java : int 4byte, long 8 byte

# 타입 변경
series = series.astype("int32")
series
```

```python
# 제거

del series['미술']
series
```
