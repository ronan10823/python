#### apply()


```python
df = pd.DataFrame({"yyyy-mm-dd":['2005-09-28','2007-10-05','2012-12-20']})
df
```

```python
def extract_year(value):
    return value.split("-")[0]
```

```python
# year 컬럼 생성

# Directely Create X
# df['year']=['2005']

df['year'] = df['yyyy-mm-dd'].apply(extract_year)
df
```

```python

def get_age(year, curr_year):
    return curr_year - int(year)
```

```python
# age 컬럼 생성

df['age'] = df['year'].apply(get_age, curr_year=2026)
df
```

```python
df = pd.DataFrame({"영어":[60,70],"수학":[100,50]},index=["Dave","David"])
df
```

```python
def func(df):
    # print(df.index)
    # print(df.values)
    df['영어'] = 80
    return df
```

```python
# (default) axis = 0 (apply in colums, ↓, default) 
#           axis = 1 (apply in rows, → ) 
df.apply(func, axis=0)
```

```python
# 영어, 수학을 동일하게 80점을 적용
# (일반적) 함수로 원하는 컬럼에 원하는 형태대로 적용

df2 = df.apply(func, axis=1)
df2
```
