```python
# 정규식

# \를 넣어야 맨 처음의 엔터가 빠진다. 
data = """ \
park 800905-1049118
kim 700905-1059118
"""

# 뒷자리를 ****** 출력

# ver 1.0 정규식 사용X
# 엔터 기준으로 문자열 분리
words = data.split("\n")
print(words)

# 결과를 담을 리스트 생성
result = []

# 공백을 기준으로 문자열 분리
for line in words:
    word_result = []
    # person = line.strip().split(" ")

    for word in line.strip().split(" "):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6] + "-" + "*******"
        word_result.append(word)
    result.append(" ".join(word_result))

# print(result) 리스트(list)

#  list -> 문자열로 변경
print("\n".join(result))
```

#### 정규식
- 1. 문자열이 가지고 있는 메소드 사용
- 2. re 모듈 
    - compile() : 패턴 생성
    - search() : 문자열 전체를 검색해서 정규식과 매칭되는 패턴을 찾아서 리턴(Match 객체)
    - sub() : 찾아서 변경 
    - findall() : 패턴과 일치하는 모든 문자열을 찾아서 리스트로 반환

```python
# train 읽어서 남성, 미혼여성, 기혼여성, 기타 각각의 시트로 저장
# 보고서 시트 생성
# 분류  생존자수  사망자수  생존률
# 남성    150     80       60



# csv 생성
wb = Workbook()

# 첫 번째 시트 가져오기
ws_man = wb.active
# 시트명 변경
ws_man.title = "남성"

# n번째 시트 생성
ws_solo_women = wb.create_sheet()
ws_solo_women.title = "미혼 여성"

ws_married_women = wb.create_sheet()
ws_married_women.title = "기혼 여성"

ws_others = wb.create_sheet()
ws_others.title = "기타"

# 보고서 시트
ws_report = wb.create_sheet()
ws_report.title = "보고서"

# 생존자수, 사망자수 변수 선언
man_survived, man_unsurvived = 0, 0
solo_woman_survived, solo_woman_unsurvived = 0, 0
married_woman_survived, married_woman_unsurvived = 0, 0
others_survived, others_unsurvived = 0, 0

# 남자( Mr.\ ), 미혼여성( Miss.\ ), 기혼여성( Ms.\ ), 기타
pattern = re.compile(r" [A-Za-z]+\.")

# 엑셀 데이터 가져오기
train_wb = load_workbook("./data/train.xlsx")
train_ws = train_wb.active

for row in train_ws.iter_rows(max_row=30):
    # 이름 컬럼
    result = pattern.search(row[3].value)
    # 생존 여부 컬럼
    survived = row[1].value

    # 제목 행 붙여넣기
    if row[0].row == 1:
        ws_man.append([item.value for item in row])
        ws_married_women.append([item.value for item in row])
        ws_solo_women.append([item.value for item in row])
        ws_others.append([item.value for item in row])
        
    else:
        if result:
            name = result.group()
            if name.strip() == "Mr.":
                ws_man.append([item.value for item in row])                 
                # survied 0인 경우 남자 사망자수 +=1
                if survived == 0:
                    man_unsurvived +=1
                else:
                # survied 1인 경우 남자 생존자수 +=1
                    man_survived +=1
            elif name.strip() == "Miss.":
                ws_married_women.append([item.value for item in row])
                if survived == 0:
                    married_woman_unsurvived +=1
                else:
                    married_woman_survived +=1
            elif name.strip() == "Mrs.":
                ws_solo_women.append([item.value for item in row])
                if survived == 0:
                    solo_woman_unsurvived +=1
                else:
                    solo_woman_survived +=1
            else:
                ws_others.append([item.value for item in row])
                if survived == 0:
                    others_unsurvived +=1
                else:
                    others_survived +=1

# 보고서 작성
ws_report.append(["분류","생존자수","사망자수","생존률"])

rate = "%.f2%%" % ( (man_survived / (man_survived + man_unsurvived))* 100) 
ws_report.append(["남자",man_survived,man_unsurvived,rate])

rate = "%.f2%%" % ( (married_woman_survived / (married_woman_survived + married_woman_unsurvived))* 100) 
ws_report.append(["기혼여성",married_woman_survived,married_woman_unsurvived,rate])

rate = "%.f2%%" % ( (solo_woman_survived / (solo_woman_survived + solo_woman_unsurvived))* 100) 
ws_report.append(["미혼여성",solo_woman_survived,solo_woman_unsurvived,rate])

rate = "%.f2%%" % ( (others_survived / (others_survived + others_unsurvived))* 100) 
ws_report.append(["기타",others_survived,others_unsurvived,rate])


wb.save("./data/file1.xlsx")
wb.close()
```
