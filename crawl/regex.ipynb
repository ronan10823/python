#### 정규표현식
- 정규식 패턴은 동일

```python
# 정규식

# \를 넣어야 맨 처음의 엔터가 빠진다. 
data = """ \
park 800905-1049118
kim 700905-1059118
"""

# 뒷자리를 ****** 출력

# ver 1.0 정규식 사용X
# 엔터 기준으로 문자열 분리
words = data.split("\n")
print(words)

# 결과를 담을 리스트 생성
result = []

# 공백을 기준으로 문자열 분리
for line in words:
    word_result = []
    # person = line.strip().split(" ")

    for word in line.strip().split(" "):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6] + "-" + "*******"
        word_result.append(word)
    result.append(" ".join(word_result))

# print(result) 리스트(list)

#  list -> 문자열로 변경
print("\n".join(result))
```
