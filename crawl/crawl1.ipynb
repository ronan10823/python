```python
# put, post, delete => 

url = "https://httpbin.org"

s = requests.Session()

data = {"name":"hong"}

# post 세션 신청 방법 r = s.post(f"{url}/post", json)
r = s.post(f"{url}/post", data=data)
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python
s = requests.Session()

r = s.delete(f"{url}/delete")
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python
s = requests.Session()

data = {"name":"hong"}

# post 세션 신청 방법 r = s.post(f"{url}/post", json)
r = s.put(f"{url}/put", data=data)
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python
# get, parameter

s = requests.Session()

# parameter : ? 뒤의 부분
# http://www.naver.com?search=mac 
param = {"name":"hong"}

r = s.get(f"{url}/get", params=param)
print(r.ok)
print(r.text)

# 세션 종료
s.close()
```

```python
# 뉴스 기사 (동계 올림픽 관련, 100개의 데이터 가져오기)

url = "https://openapi.naver.com/v1/search/news.json"

start = 1

with requests.Session() as s:

    # 뉴스를 한 번에 100개씩 꺼내기 -> start(시작 번호)가 1, 101, 201.. 이렇게 되어야 한다. 
    # ramge(10) -> 0-9를 의미
    for idx in range(3):
        start_num = start + (idx * 100)

        param = {
            "query" : "동계 올림픽",
            "display" :"100",
            "start":str(start_num) # param에 넣으면 start_num (int)를 자동으로 str로 바꿔주기도 한다. 
        } 

    res = s.get(url, headers=headers, params=param) 
    contents = res.json()
    for news in contents['items']:
        print(news['title'])
```

```python
# book api => title, author, discount, publisher

url = "https://openapi.naver.com/v1/search/books.json"

start = 1

with requests.Session() as s:

    for idx in range(3):
        start_num = start + (idx * 100)

        param = {
            "query" : "심리",
            "display" :"100",
            "start":start_num
        } 

        res = s.get(url, headers=headers, params=param) 
        contents = res.json()
        for book in contents['items']:
            print({book['title'],book['author'],book['discount'],book['publisher']})
```

```python
with open("./data/story.html", "r") as fr:
    story = fr.read()

soup = BeautifulSoup(story, "html.parser")

# 태그
print(f"title 요소 {soup.title}")
print(f"title 요소 text {soup.title.text}")
print(f"p 요소 {soup.p}")
print(f"p 요소 text {soup.p.text}")
```

```python
# 문서 구조

title = soup.title
print(f"title 부모 요소 : {title.parent}")

# body의 자식 요소들 (모두 찾기)
children = soup.body.children
# print(children)
for child in children:
    print(child)
```

```python
# find-- (가장 많이 사용하는 함수)
# find("기준", "속성") / find_all()

print(f"title {soup.find("title")}")
print(f"p {soup.find("p")}")
print(f"p {soup.find("p", class_="story")}")
print(f"p {soup.find("a", id="link1")}")
print(f"p {soup.find("a", attrs={"class":'story', "data-num":"3"})}")
```
