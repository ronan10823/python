````python
# 가변 파라메터
# 파라메터의 개수가 일정치 않은 경우 사용

def info(*args):
    print(args)


info("wine")
info("wine","35","45","65")
info("35","45","65")
info()
````

````python
def sum_mul(choice, *args):
    if choice == "+":
        result = 0
        for i in args:
            result += i
    elif choice == "*":
        result = 1
        for i in args:
            result *= i
    return result

print(sum_mul("+", 1,2,3,4,5))
print(sum_mul("*", 1,2,3,4,5))
````

````python
## **kwargs : 딕셔너리 형태로 처리

def args_func1(**kwargs):
    for k,v in kwargs.items():
        print(f"{k} : {v}")

args_func1(name="Park")
args_func1(name="Park", name2="Kim", name3="Hong", active=True)
args_func1(name="Park", age=25, title="Title")
````

````python
def args_func2(arg1, arg2=5, *args, **kwargs):
    print(arg1, arg2, args, kwargs)

args_func2(10,20)
args_func2(10,20,"park","kim")
args_func2(10,20,"park","kim", age=14, name="choi")
````

````python
# 다중 리턴 가능 : tuple 에 담아서 보내줌

def sum_and_mul(a, b):
    return a + b, a * b


result = sum_and_mul(5,3)
print(result)
print(result[0])


sum1, mul1 = sum_and_mul(9,3)
print(f"{sum1}, {mul1}")
````

````python
def sum_and_mul(a, b):
    return [a + b, a * b]


result = sum_and_mul(5,3)
print(result)
print(result[0])


sum1, mul1 = sum_and_mul(9,3)
print(f"{sum1}, {mul1}")
````

````python
# 함수 정의 : 사칙연산 (+,-,*,//)
def fourRules(num1, num2, op):
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "//":
        return num1 // num2
    

print(fourRules(5,6,"+"))
print(fourRules(5,6,"-"))
print(fourRules(5,6,"*"))
print(fourRules(13,6,"//"))
````

````python
def fourRules(num1, num2, op):
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "//":
        return num1 // num2


# 사용자한테 두 수와 연산자 입력받기
num1 = int(input("Input Num1 : "))
num2 = int(input("Input Num2 : "))
op = input("Input operand : ")


print(f"{num1}{op}{num2}={fourRules(num1,num2,op)}")
````

````python
# 
def get_number():
    return random.randrange(1,46) # 1~45 임의의 수


# 로또번호 추출
# 리스트, set 
lotto = set()
while True:
    lotto.add(get_number())
    if len(lotto) >= 6:
        break

# 정렬
list1 = list(lotto)
list1.sort()

print("lotto ",list1)
````

````python
# 전역변수
a = 1

def func1(a):
    print("지역변수 ",a)
    a += 1
    print("덧셈 후 지역변수 ",a)

func1(5)
print(a)
````

````python
a = 1

def func1():  
    global a  # 함수 내부에서 전역변수로 선언된 변수 접근 
    a += 1
    print("덧셈 후 변수 ",a)

func1()
````

### 람다 함수
- 단일문으로 표현되는 익명 함수
- 1회성으로 함수를 만들어 사용

````python
# 숫자를 받아서 **2 리턴

def square(num):
    return num ** 2

print(square(5))
````

````python
square = lambda num:num ** 2
square(6)
````

````python
add = lambda x,y:x+y
add(6,9)
````

````python
# filter, map, reduce

list1 = [1,2,3,6,8,9]

# 짝수 요소만 추출한 후 새로운 리스트 생성하기
evens = []

for i in list1:
    if i % 2 == 0:
        evens.append(i)

evens
````

````python
evens = [ i for i in list1 if i % 2 == 0]
evens
````

````python
evens = []

def even(list1):
    for i in list1:
        if i % 2 == 0:
            evens.append(i)

even([1,2,3,6,8,9])
print(evens)
even([11,22,3,6,8,9,12,14])
````
