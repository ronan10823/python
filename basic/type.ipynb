```python
# 비밀번호 생성

# http://naver.com
# 규칙 1 : http:// 제외 => naver.com
# 규칙 2 : 첫번째로 나오는 마침표 이후 부분은 제외 => naver
# 규칙 3 : 남은 글자 중 처음 세 자리 + 글자 갯수 + 글자 내 'e' 문자 갯수 + ! => nav51!

# "123" + 3 => TypeError
# "123" + str(3) => TypeError 해결

url = "http://naver.com"
password = url.replace("http://","") # url[7:]
password = password[:password.find(".")]

e_count = password.count("e")
length = len(password)

password = password[:3] + str(length) + str(e_count) + "!"
print(password)
```

```python
classA = [70, 60, 55, 75, 95, 90, 80, 85, 100]

# classA 점수의 합, 평균 구하기
total = 0
for num in classA:
    total += num

print(f"classA 점수 합계 : {total}, 평균 : {total/len(classA)}")

# sum()
print(f"classA 점수 합계 : {sum(classA)}, 평균 : {sum(classA)/len(classA)}")

total = sum(classA)
avg = total/len(classA)
print(total, avg)
```

```python
# 주차장 프로그램
# 5대만 주차 가능 (차례대로 입차 / 출차는 반대 순서)

parking_lot = []
top, car_name = 0, "A"
# len을 대신할 변수, top

while True: 
    no = int(input("[1] 자동차 넣기 | [2] 자동차 빼기 | [3] 종료 "))
    if no <=3 :
        if no == 1:

            if top >=5:
                print("주차장 만차")
            else:
                parking_lot.append(car_name)
                print(f"{car_name} 자동차 들어감. 주차장 상태 ==> {parking_lot}")
                top += 1
                # car_name 새로 구하기
                car_name = chr(ord(car_name)+1) 
            # if len(parking_lot)>=5 :
            #     print("만차")
            #     break
            # else:
            #     car_name = chr(ord(car_name)+1)
                # print(f"{car_name} 자동차 들어감. 주차장 상태 ==> {parking_lot}")

        elif no == 2:
            if top > 0:
                # 마지막 꺼내기 : pop()
                outCar = parking_lot.pop()
                print(f"{outCar} 자동차 빠짐. 주차장 상태 ==> {parking_lot}")
                
                top -= 1

                # car_name 새로 구하기
                car_name = chr(ord(car_name)-1)
            else:
                print("빠져나갈 자동차 없음")
            
        elif no == 3:
            print("프로그램 종료")
            break # 프로그램 종료
    else: 
        print("메뉴 번호를 확인해 주세요")
```
