```python
# sort( {필드명 : 1}) : 오름차순
# sort( {필드명 : -1}) : 내림차순

# find({},{"title":1,"_id":0, "year":1} ) : 전체 조회하되, 컬럼명은 지정한 것만 보여주기 (_id는 무조건 나오기 때문에 무조건 0으로 지정)
docs = db.movies.find({},{"title":1,"_id":0, "year":1} ).sort({"imdb.rating":1}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 상영 시간이 100분 이상인 영화 조회 (제목, 상영시간, 등급)

docs = db.movies.find({"runtime":{"$gte":100}},{"title":1, "runtime":1, "rated":1, "_id":0}).limit(30)
for doc in docs:
    print(doc)
```

```python
# imdb 평점이 8점 이상인 영화 조회 (제목, imdb) 10개만

docs = db.movies.find({"imdb.rating":{"$gte":8}},{"title":1, "imdb":1, "_id":0}).limit(10)
for doc in docs:
    print(doc)
```

```python
# imdb 평점이 8점 이상이고, 2000년 이후에 개봉한 영화 조회 (제목, imdb) 10개만

docs = db.movies.find({"imdb.rating":{"$gte":8}, "year":{"$gte":2000}},{"title":1, "imdb":1, "_id":0}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 제목에 love(대소문자 구분없이)가 들어가는 영화 (plot, title 10개만)

docs = db.movies.find({"title":{"$regex":"love", "$options":"i"}}, {"title":1, "plot":1, "_id":0}).limit(10)
for doc in docs:
    print(doc)
```

```python
# 제작 연도가 1920년 이전인 영화 개수

# docs = db.movies.find({"year":{"$lt":1920}})
# len(docs.to_list())

db.movies.count_documents({"year":{"$lt":1920}})
```

```python
# 장르가 Comedy, Romance 둘 다 포함된 영화, (title, genres, 10개)
# in : or
# $all : 모두

docs = db.movies.find({"genres":{"$all":["Comedy", "Romance"]}})
for doc in docs:
    print(doc)
```

```python
# 최신 영화 10개(sort, year), (title, genres, 20개)

docs = db.movies.find({}, {"title":1, "genres":1, "_id":0}).sort({"year":1}).limit(20)
for doc in docs:
    print(doc)
```

```python
# 영화 제작년도가 1990 - 1999 년 사이의 영화 조회 (title, year, 10개)

docs = db.movies.find({"year":{"$gte":1990, "$lte":1999}}, {"title":1,"year":1, "genres":1, "_id":0}).limit(10)
for doc in docs:
    print(doc['year'], doc['title'])
```

```python
# 모든 장르 보기

print(db.movies.distinct("genres"))
```

```python
# 연도별 영화 개수

rows = db.movies.aggregate([
                            {"$group":{"_id":"year", "count":{"$sum":1}}}
                            ,{"$sort":{"_id":-1}}
                            ,{"$limit":10}
                            ])

for row in rows:
    print(row)
```

```python
# 장르별 영화 개수

rows = db.movies.aggregate([
                            {"$unwind":"$genres"}
                            ,{"$group":{"_id":"$genres", "count":{"$sum":1}}}
                            ,{"$sort":{"_id":-1}}
                            ,{"$limit":10}
                            ])

for row in rows:
    print(row)
```

```python
# imdb 평점 평균이 높은 감독
# 감독을 그룹으로 잡아야 한다.

rows = db.movies.aggregate([
                    {"$unwind":"$directors"},
                    {"$group":{"_id":"$directors", "avgRating":{"$avg":"$imdb.rating"}, "count":{"$sum":1}}},
                    {"$match":{"count":{"$gte":5}}},
                    {"$sort":{"avgRating":-1}},
                    {"$limit":10}
                    ])

for row in rows:
    print(row)
```
