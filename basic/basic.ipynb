#### 삼항 연산자

- python : ( 조건 ? 참 : 거짓 ) 형태의 삼항 연산자가 없다.
- python 삼항연산자 형태 : 참_값 if 조건 else 거짓_값

```python
# print(a<100 ? "a는 100보다 작다" : "a는 100보다 작다" )

result = "a는 100보다 작다" if a < 100 else "a는 100보다 크다"
print(result)
```

#### 다중 조건
- java: if (condition...) else if (condition2...)
- python : if condition... : elif condition...: else

```python
num = 90

if num >= 90:
    print("A")
elif num >= 80:
    print("B")
elif num >= 70:
    print("C")
elif num >= 60:
    print("D")
else: 
    print("F")
```

```python
# 사칙 연산기 (숫자 2, 연산자 => +, -, *, /, //, %)

num1 = int(input("1번째 숫자 입력 : "))
num2 = int(input("2번째 숫자 입력 : "))
oper = input("사칙연산 +, -, /, *, //, % 중 입력")

if oper=="+":
    result = (num1 + num2)
elif oper=="-":
    result = (num1 - num2)
elif oper=="/":
    result = (num1 // num2)
elif oper=="//":
    result = (num1 // num2)
elif oper=="*":
    result = (num1 * num2)
elif oper=="%":
    result = (num1 % num2)

print(f"{num1} {oper} {num2} = {result}")

match oper:
    case "+":
        print(num1 + num2)
    case "-":
        print(num1 - num2)
    case "/":
        print(num1 / num2)
    case "*":
        print(num1 * num2)
    case "//":
        print(num1 // num2)
    case "%":
        print(num1 % num2)
```
