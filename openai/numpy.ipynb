#### 넘파이(Numpy)

```python
import numpy as np
```

```python
# 1차원 벡터 생성
array1 = np.array([1,2,3])
array1
```

```python
print(type(array1))
print(array1.shape) # (3, ) : 3열 (1차원 배열이기에 행열 정보 없음)
```

```python
# 2차원

array2 = np.array([[1,2,3],[4,5,6]])
array2
```

```python
print(type(array2))
print(array2.shape) # (2,3) 2행 3열
```

```python
array2 = np.array([[1,2,3]])
print(array2)
print(array2.shape) # (1,3) 1행 3열
```

```python
# 파이썬 리스트
list1 = [1, 2, 'a', 'b', True]
list1
```

```python
array1.dtype # int64 ( int, byte )
```

```python
array3 = np.array([1,2,3.1])
print(array3.dtype) # float64 -> float로 잡는다. 
```

```python
array4 = np.array(list1)
print(array4.dtype) # <U21
```

- ndarray 다양한 방법으로 생성하기

```python
# arange(숫자) / arange(시작값, 숫자)
np.arange(10)
np.arange(5, 10)
np.arange(1,10,2) # array([1, 3, 4, 7, 9])
```

```python
# zeros 0으로 채우기
# dtype 지정 안 하는 경우 float
# int32 / int64 -> int64 = 8 bite(사용용량) 사용
np.zeros(5, dtype="int32")
```

```python
# ones 1로 채우기
np.ones(5, dtype="int32")
```

```python
np.ones((2,2))
```

```python
np.ones((3,2,2))
```

```python
# full - 특정 숫자로 채우기
np.full((3,4),7)
```

```python
# eye - 단위행렬
np.eye(5)
```

```python
# linspace - 처음값(고정) - 끝값(고정) 사이에 랜덤한 숫자 출력
np.linspace(1,10,5)
```

- 차원 변경 : reshape()
- 1차원 <--> 2차원

```python
# 1차원 (1 * 10)
array1 = np.arange(10)
print(array1)
```

```python
# 1 -> 2 (5 * 2)
array1.reshape(5,2) # 개수가 맞아야 한다. 
```

```python
# 열(5)만 지정, 행은 알아서 지정
array1.reshape(-1, 5)
```

```python
array1.reshape(5,-1)
```

```python
# 임의의 ndarray : 정규 분포

np.random.randn(2,3)
```

```python
# 행렬 사칙연산

x = np.array([[4,4,4],[8,8,8]])
y = np.array([[1,1,1],[2,2,2]])
```

```python
x + y
```

```python
x - y
```

```python
x * y
```

```python
x // y
```

```python
x * 10
```

```python
# 행렬 내접의 곱

x = np.array([[1,2,3],[4,5,6]])
y = np.array([[7,8],[9,10],[11,12]])
print(x)
print(y)
```

```python
# dot으로 행렬 연산자 곱 -> ? 어떻게?
np.dot(x,y)
```

```python
# 전치행렬(행렬 뒤바꾸기)

np.transpose(x)
```

```python
v = np.array([3,4])

# norm(ord = Norm의 종류)
# L1(Manhattan distnace), L2(Euclidean distance), Max(절댓값 중 가장 큰 값)
# (Default/Common) L2
np.linalg.norm(v)
```
