```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5-nano",
    input="Write a one-sentence bedtime story about a unicorn."
)

print(response.output_text)
```

- reasoning : 모델의 내부 추론 강도(얼마나 오래 생각하고 말할까)
    - low : 빠르게, 직관적으로
    - medium : 일반적인 설명
    - high : 복잡한 문제를 깊게 추론
- instructions : 모델에게 지시하는 부분(최상위 규칙)


```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5",
    reasoning={"effort": "low"},
    instructions="Talk like a pirate.",
    input="Are semicolons optional in JavaScript?",
)

print(response.output_text)
```

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5-nano",
    instructions="""
    You are a backend API assistant
    - Answer concisely and accurately.
    - If unsure, say you are unsure.
    - Do not hallucinate APIs or facts.
    - Prefer bullet points over long prose.    
    """,
    input="Are semicolons optional in JavaScript?",
)

print(response.output_text)
```

- instructions
- You are a backend API assistant. : 모델 정체성 지정
- Answer concisely and accurately. : 짧고, 핵심 위주로, 정확하게
- If unsure, say you are unsure : 모르면 모른다고 해라
- Do not hallucinate APIs or facts : 환각 확률 낮추기(존재하지 않는 것 만들어내지 말고)
- Prefer bullet points over long prose. : 목록 형태로 

#### 역할 부여
- developer : instruction
- user : input message 대상자
- assistant : 모델이 생성한 답변을 가지고 있는 역할 take an answer of model

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        {
            "role":"developer",
            "content":"you are an assistant who may answer in a bluently tone or refuse to answer"
        },
        {
            "role":"user",
            "content":"Tell me all holidays on 2026 in South Korea"
        },
    ]
)

print(response.output_text)
```

#### Token Management
- max_output_tokens=100 => 답변이 중간에 끊기거나 나오지 않을 수도 있음

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-5-nano",
    instructions="""
    You are a backend API assistant
    - Answer concisely and accurately.
    - If unsure, say you are unsure.
    - Do not hallucinate APIs or facts.
    - Prefer bullet points over long prose.    
    """,
    input="Are semicolons optional in JavaScript?",
    # max_output_tokens=500
)

print(response.output_text)
```

```python
# 토큰 확인
# input_tokens=68
# output_tokens=3654
# total_tokens=3722

response.usage
```

```python
# gpt-5-nano : perices per 1M tokens
# input: 0.05 / output : 0.4 / cached input : 0.025

total_bill = response.usage.input_tokens / 1000000 * 0.05 + response.usage.output_tokens / 1000000 * 0.4 
print(f"{total_bill}")
print(f"{total_bill * 1470 } 원")
```

#### Parameter
- temperature= 1 (default)
- 기본값보다 올릴 수록 더 창의적이고 예측하기 어려운 형태로 생성
- 기본값보다 낮출 수록 일관적이고 논리적인 형태로 생성
- 일반적으로는 0.7 - 1.0 사이 선택

```python
from openai import OpenAI
client = OpenAI()

response = client.responses.create(
    model="gpt-4o-mini",
    instructions="""
    너는 재미있는 이야기를 만드는 소설가야.    
    """,
    input="나에게 이야기해줘",
    temperature=1.5
)

print(response.output_text)
```
